```python
import matplotlib.pyplot as plt
```

```python
containers = [10 * i for i in [10, 80, 31, 37, 17, 20, 90, 73, 50, 89]]
people = [1, 6, 2, 3, 1, 2, 10, 4, 4, 8]
profit = [[containers[i] / (people[i] + j) for i in range(len(containers))] for j in range(5)]
for addedP in profit:
    print([round(i * 100) / 100 for i in addedP])
```

```python
fig, ax = plt.subplots()
x = [y for y in range(len(profit))]
for i in range(len(profit[0])):
    w = [profit[y][i] for y in x]
    ax.plot(x, w)
plt.show()
```
